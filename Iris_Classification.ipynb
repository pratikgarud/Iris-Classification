```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
```

```python
iris = pd.read_csv('iris.csv')
iris.head()
```

```python
sns.lmplot(x='sepal_length',y='sepal_width',data=iris)
```

```python
sns.lmplot(x='petal_length',y='petal_width',data=iris)
```

```python
sns.residplot(x=iris['sepal_length'],y=iris['sepal_width'],color='red')
```

```python
sns.residplot(x=iris['petal_length'],y=iris['petal_width'],color='blue')
```

```python
sns.regplot(x=iris['sepal_length'], y=iris['sepal_width'], data=iris, color='red', label='order 1', order=1)
```

```python
sns.regplot(x=iris['sepal_length'], y=iris['sepal_width'], data=iris, color='blue', label='order 2', order=2)
```

```python
sns.regplot(x=iris['petal_length'], y=iris['petal_width'], data=iris, color='green', label='order 3', order=3)
```

```python
sns.lmplot(x='sepal_length', y='sepal_width', data=iris, hue='species' ,palette='Set1')
```

```python
X = iris[['sepal_length','sepal_width','petal_length','petal_width']]
```

```python
y = iris['species']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
model.predict(X_test)
```

```python
import pickle
pickle.dump(model,open('model.pkl','wb'))
```
